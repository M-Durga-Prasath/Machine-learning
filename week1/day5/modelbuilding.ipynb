```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
# data collection an sorting
data = pd.read_csv("../../data/titanic/titanic.csv")
data.shape
data.columns
data.head()
```

```python

features = ['Fare', 'Pclass', 'Sex', 'Age']
x = data[features]
y = data['Survived']
y
x['Sex'] = x['Sex'].map({'female': 0, 'male': 1})
```

```python
# splitting data into Train, Validation, Test

x_train,x_temp,y_train,y_temp = train_test_split(
    x,
    y,
    test_size = 0.25,
    random_state=42
)

# splitting temp i.e testing datat into validation and testing data
x_val, x_test, y_val, y_test = train_test_split(
    x_temp,
    y_temp,
    test_size=0.5,
    random_state=42
)
x_train
```

```python
# cleaning handling msising values
train_median = x_train["Age"].median() #to prevent data leakage not req at this level but useful
x_train['Age'] = x_train['Age'].fillna(train_median)
x_val['Age'] = x_val['Age'].fillna(train_median)
x_test['Age'] = x_test['Age'].fillna(train_median)
x_train
```

```python
# kyu kar re hai nai malum waps gpt mar le
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_val = scaler.transform(x_val)

x_test = scaler.transform(x_test)
x_train
```

```python
# model 1 - Logistic Regression
Lmodel = LogisticRegression()
Lmodel.fit(x_train, y_train)
lreg_score = Lmodel.score(x_val, y_val)
lreg_score
```

```python
# Model 2 - randomforest
rmodel = RandomForestClassifier()
rmodel.fit(x_train, y_train)
rand_score = rmodel.score(x_val,y_val)
rand_score
```

```python
# Model 3 - gradientboosting
gradmodel = GradientBoostingClassifier()
gradmodel.fit(x_train, y_train)
grad_score = gradmodel.score(x_val,y_val)
grad_score
```

```python
gradmodel.score(x_test,y_test)
```
