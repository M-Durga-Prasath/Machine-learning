```python
# numpy_basics.py

import numpy as np

a = np.array([1, 2, 3])
b = np.array([[4, 5, 6], [7, 8, 9]])

print("Addition:", a + b)
print("Addition:", a - b)
print("Multiplication:", a * b)
print("Mean:", np.mean(a))
print("Shape:", a.shape)
```

```python
arr = np.ones((5,5),dtype="int8")
arr[1:4,1:-1] = [0,0,0]
arr[2,2] = 9
arr
```

```python
arr = np.ones((6,5),dtype="int8")
c = 0
for i in range(1,31):
    arr[c,(i-1)%5] = i
    if i%5 == 0:
        c += 1
arr
```
