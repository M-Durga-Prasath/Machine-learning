```python
class Model(nn.Module):
    def __init__(self,input = 4, h1=10, h2 = 10, out = 3):
        super().__init__()
        self.forward1 = nn.Linear(input,h1)
        self.forward2 = nn.Linear(h1,h2)
        self.output = nn.Linear(h2,out)
    
    def forward(self, x):
        x = nn.functional.relu(self.forward1(x))
        x = nn.functional.relu(self.forward2(x))
        x = nn.functional.relu(self.output(x))
        return x
    
```

```python
model = Model()

data = load_iris()
x = data.data
y  = data.target
# x
# x
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train = torch.FloatTensor(x_train)
x_test = torch.FloatTensor(x_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
# x_train
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3) #1e-3 matlab 0.0001
# model.parameters
```

```python
losses = []
for epoch in range(100):
    optimizer.zero_grad()          # clear old gradients
    output = model.forward(x_train)        # forward pass
    loss = loss_fn(output, y_train)         # compute loss
    loss.backward()                # backprop
    optimizer.step() 
    if epoch%10 == 0:
        print(loss)
        losses.append(loss)
```

```python
with torch.no_grad():
    y_eval = model.forward(x_test)
    loss = loss_fn(y_eval, y_test)
loss
```
