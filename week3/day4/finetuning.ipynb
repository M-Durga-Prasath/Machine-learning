<!-- commone thinsg -->
| Metric                  | What it Measures                         | Better Value                                                     |
| ----------------------- | ---------------------------------------- | ---------------------------------------------------------------- |
| **best.pt**             | Best-performing saved model              | Use for inference                                                |
| **last.pt**             | Final model after training               | Useful for resuming training                                     |
| **Box Loss**            | Bounding box accuracy                    | Lower ↓                                                          |
| **Classification Loss** | Class prediction error                   | Lower ↓                                                          |
| **DFL Loss**            | Bounding box edge precision              | Lower ↓                                                          |
| **Precision**           | Correct detections among all detections  | Higher ↑                                                         |
| **Recall**              | Detected objects among all real objects  | Higher ↑                                                         |
| **mAP50**               | Detection accuracy at IoU ≥ 0.5          | Higher ↑                                                         |
| **mAP50-95**            | Detection accuracy across IoU 0.50–0.95  | Higher ↑                                                         |
| **IoU**                 | Overlap between predicted and true boxes | Higher ↑                                                         |
| **Confidence Score**    | Model's certainty about a detection      | Higher is more confident (threshold chosen based on application) |
| **True Positive (TP)**  | Correct detection                        | More ↑                                                           |
| **False Positive (FP)** | Incorrect detection                      | Fewer ↓                                                          |
| **False Negative (FN)** | Missed detection                         | Fewer ↓                                                          |
| **Learning Rate**       | Size of weight updates                   | Depends on training setup                                        |
| **Epoch**               | One full pass through the dataset        | More isn't always better                                         |
| **Batch Size**          | Images processed per optimization step   | Depends on hardware and model                                    |
| **Training Loss**       | Error on training data                   | Lower ↓                                                          |
| **Validation Loss**     | Error on validation data                 | Lower ↓                                                          |
